```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python


# Read the CSV file into a pandas DataFrame
df = pd.('movies.csv')

# Split the genres column into a list of genres
df['genres'] = df['genres'].str.split('|')

# Create a list of all unique genres
genres_list = list(set([genre for genres in df['genres'] for genre in genres]))

# Calculate the frequency of each genre
genre_counts = {genre: 0 for genre in genres_list}
for genres in df['genres']:
    for genre in genres:
        genre_counts[genre] += 1

# Convert the dictionary to a pandas DataFrame and sort by frequency
genre_counts_df = pd.DataFrame.from_dict(genre_counts, orient='index', columns=['count'])
genre_counts_df = genre_counts_df.sort_values(by='count', ascending=False)

# Plot the bar chart
genre_counts_df.plot(kind='bar', figsize=(12, 5))

plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Movie Genres')
plt.savefig('genre_distribution.png')

```
